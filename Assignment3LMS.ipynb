```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')

print("Sample Data:")
print(df.head())

print("\nColumns:")
print(list(df.columns))

print("\nDataset Info:")
print(df.info())
print("\nStatistical Summary:")
print(df.describe())
```

```python
df.plot(x='sepal_length', y='sepal_width', kind='line', title='Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(df['petal_length'], kde=True, color='green')
plt.title('Histogram: Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['species'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Species Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Boxplot: Sepal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title('Scatter: Sepal Length vs Petal Length')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Boxplot: Sepal Length by Species')
plt.show()
```
